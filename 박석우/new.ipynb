```python
def cat_pc(x):
    if x in ['vrf', 'single-split', 'chiller', 'heating', 'multi-split',
             'rac',]:
        return 'heating'
    elif x in ['ventilation']:
        return 'ventilation' 
    elif x in ['control']:
        return 'control'
    elif x in ['interactive signage', 'video wall signage', 'standard signage', 
               'high brightness signage', 
               'one:quick series', 'digital signage']:
        return 'digital_signage'
    elif x in ['led signage']:
        return 'led_signage'
    elif x in ['hotel tv', 'hospital tv', 'special signage', 'accessories',
               'pro:centric', 'commercial tv']:
        return 'commercial_tv' ## digital_signage에 속하긴 함.
    elif x in ['oled signage']:
        return 'oled_signage'
    elif x in ['monitior']:
        return 'monitors'
    elif x in ['medical display']:
        return 'medical_display'
    elif x in ['cloud device']:
        return 'cloud_device'
    elif x in ['laptop', 'notebook']:
        return 'laptops'
    elif x in ['webos', 'software solution', 'signage care solution', 'software',
               'technical support ', ]:
        return 'software/service'
    elif x in ['projector']:
        return 'projectors'
    elif x in ['ess']:
        return 'ess'
df_train['product_category']=df_train['product_category'].apply(cat_pc)
df_test['product_category']=df_test['product_category'].apply(cat_pc)
```
