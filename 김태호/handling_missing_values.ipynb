```python
# DataWig SimpleImputer 초기화
imputer = datawig.SimpleImputer(
    input_columns=['customer_country', 'business_unit'], # 결측치 예측에 사용될 특성 열
    output_column='product_category',  # 결측치를 채울 대상 열
    output_path='imputer_model'  # 모델 파일 저장 경로
)
```

```python
def fill_missing_values(df, column_to_impute, columns_to_use):
    """
    df: DataFrame with missing values.
    column_to_impute: The column for which the missing values should be imputed.
    columns_to_use: A list of column names to be used as input features for imputing.
    """
    # Initialize a SimpleImputer model
    imputer = datawig.SimpleImputer(
        input_columns=columns_to_use,
        output_column=column_to_impute
    )

    # Fit the imputer model on the complete data
    imputer.fit(train_df=df, num_epochs=50)

    # Impute missing values and return original dataframe with imputed values
    imputed = imputer.predict(df)

    return imputed
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from keras.models import Model
from keras.layers import Input, Dense
from keras.callbacks import EarlyStopping
import pandas as pd
import numpy as np

# 데이터 준비
df_numeric = df_train_ms.select_dtypes(include=[np.number])
df_numeric_filled = df_numeric.fillna(df_numeric.mean())  # 결측치를 평균으로 대체

# 데이터 정규화
scaler = MinMaxScaler()
df_numeric_scaled = scaler.fit_transform(df_numeric_filled)

# 데이터 분할
X_train, X_test = train_test_split(df_numeric_scaled, test_size=0.2, random_state=42)

# 오토인코더 모델 설계
input_dim = X_train.shape[1]
input_layer = Input(shape=(input_dim,))
encoded = Dense(128, activation='relu')(input_layer)
encoded = Dense(64, activation='relu')(encoded)
decoded = Dense(64, activation='relu')(encoded)
decoded = Dense(input_dim, activation='sigmoid')(decoded)

autoencoder = Model(input_layer, decoded)
autoencoder.compile(optimizer='adam', loss='mean_squared_error')

# 모델 학습
early_stopping = EarlyStopping(monitor='val_loss', patience=5, mode='min')
autoencoder.fit(X_train, X_train,
                epochs=100,
                batch_size=32,
                shuffle=True,
                validation_data=(X_test, X_test),
                callbacks=[early_stopping])

# 결측치 예측 및 대체
# 원본 데이터에서 결측치가 있던 위치
mask = df_numeric.isnull()

# 정규화된 데이터에 대해 결측치 예측
predicted = autoencoder.predict(df_numeric_scaled)

# 예측된 값 역정규화
predicted_inverse = scaler.inverse_transform(predicted)

# 역정규화된 예측값을 원본 데이터의 결측치 위치에 대체
df_numeric_filled[mask] = predicted_inverse[mask]

# 결과 확인
df_test_ms_filled = df_train_ms.copy()
df_test_ms_filled[df_numeric.columns] = df_numeric_filled

df_test_ms_filled.head()
```
